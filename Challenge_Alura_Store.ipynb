# 6. Análise Geográfica


```python
# Mapeando as regiões do BR com base na sua UF
regioes = {
    'Norte': ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO'],
    'Nordeste': ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE'],
    'Centro-Oeste': ['DF', 'GO', 'MT', 'MS'],
    'Sudeste': ['ES', 'MG', 'RJ', 'SP'],
    'Sul': ['PR', 'RS', 'SC']
}



# Criando a coluna região para cada um dos DFs usando uma função
def mapear_regiao(uf):
    if uf in ['AC', 'AP', 'AM', 'PA', 'RO', 'RR', 'TO']:
        return 'Norte'
    elif uf in ['AL', 'BA', 'CE', 'MA', 'PB', 'PE', 'PI', 'RN', 'SE']:
        return 'Nordeste'
    elif uf in ['DF', 'GO', 'MT', 'MS']:
        return 'Centro-Oeste'
    elif uf in ['ES', 'MG', 'RJ', 'SP']:
        return 'Sudeste'
    elif uf in ['PR', 'RS', 'SC']:
        return 'Sul'
    else:
        return 'Desconhecido'

loja['regiao'] = loja['local_da_compra'].apply(mapear_regiao)
loja2['regiao'] = loja2['local_da_compra'].apply(mapear_regiao)
loja3['regiao'] = loja3['local_da_compra'].apply(mapear_regiao)
loja4['regiao'] = loja4['local_da_compra'].apply(mapear_regiao)
```

```python
print("Loja 1:")
loja.groupby('regiao').agg({
    'preco': 'mean',
    'frete': 'mean',
    'avaliacao_da_compra': 'mean'
}).rename(columns={
    'preco': 'Faturamento Médio',
    'frete': 'Frete Médio',
    'avaliacao_da_compra': 'Avaliação Média'
}).round(2)
```

```python
print("Loja 2:")
loja2.groupby('regiao').agg({
    'preco': 'mean',
    'frete': 'mean',
    'avaliacao_da_compra': 'mean'
}).rename(columns={
    'preco': 'Faturamento Médio',
    'frete': 'Frete Médio',
    'avaliacao_da_compra': 'Avaliação Média'
}).round(2)
```

```python
print("Loja 3:")
loja3.groupby('regiao').agg({
    'preco': 'mean',
    'frete': 'mean',
    'avaliacao_da_compra': 'mean'
}).rename(columns={
    'preco': 'Faturamento Médio',
    'frete': 'Frete Médio',
    'avaliacao_da_compra': 'Avaliação Média'
}).round(2)
```
